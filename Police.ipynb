```python
import pandas as pd
from datetime import datetime

df = pd.read_csv('traffic_stops.csv')
```

```python
#Clearing unwanted columns
df.drop(columns=['driver_age_raw'],inplace=True)
df.drop(columns=['driver_race'],inplace=True)
df.drop(columns=['violation_raw'],inplace=True)
```

```python
df['timestamp']=pd.to_datetime(df['stop_date']+' '+df['stop_time'])
```

```python
df.fillna({'search_type':'None'},inplace=True)
```

```python
# connecting python and mysql
import pymysql
conn = pymysql.connect(
    host='localhost',
    user='root', 
    password='Mysql@1013')
cursor = conn.cursor()
cursor.execute("create database policestop_logs")

```

```python
cursor.execute("USE policestop_logs")
```

```python
cursor.execute("""create table traffic_analysis(
Vehicle_ID  int primary key auto_increment,
stop_date date,
stop_time time,
country_name varchar(100),
driver_gender varchar(10) ,
driver_age INT,
violation varchar(100),
search_conducted boolean,
search_type varchar(100),
stop_outcome varchar(100),
is_arrested boolean,
stop_duration varchar(50),
drugs_related_stop varchar(10),
vehicle_number varchar(100),
timestamp Datetime
)""")
```

```python
#Converting stop_date and stop_time to correct datetime format
from datetime import datetime
df['stop_date']=pd.to_datetime(df['stop_date'],errors='coerce',format="%Y-%m-%d")
df['stop_time']=pd.to_datetime(df['stop_time'],errors='coerce',format="%H:%M:%S")
```

```python
df
```

```python
df.dtypes
```

```python
#Insert query 

insert_query=""" insert into traffic_analysis(
stop_date,stop_time,country_name,driver_gender,driver_age,
violation,search_conducted,search_type,
stop_outcome,is_arrested,stop_duration,drugs_related_stop,
vehicle_number,timestamp) values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
```

```python
for _, row in df.iterrows():
    cursor.execute(insert_query, (
        row['stop_date'],
        row['stop_time'],
        row['country_name'],
        row['driver_gender'],
        int(row['driver_age']) if pd.notna(row['driver_age']) else None,
        row['violation'],
        bool(row['search_conducted']) if pd.notna(row['search_conducted']) else None,
        row['search_type'],
        row['stop_outcome'],
        bool(row['is_arrested']) if pd.notna(row['is_arrested']) else None,
        row['stop_duration'],
        row['drugs_related_stop'],
        row['vehicle_number'],
        row['timestamp']
    ))

    conn.commit()
```
